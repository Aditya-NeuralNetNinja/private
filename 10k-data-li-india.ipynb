```python
import json
import pandas as pd

# Read the contents of the .txt file
with open('/kaggle/input/10k-data-linkedin-india/10k_data_li_india.txt', 'r') as file:
    json_data = file.readlines()

# Initialize empty lists for job_designation and skills
job_designation_list = []
skills_list = []

# Parse each JSON entry and extract job_designation and skills
for entry in json_data:
    profile = json.loads(entry)
    job_designation = profile.get('occupation') or profile.get('headline')  # Adjust based on your JSON structure

    # Extract only the generic job designation (without company name)
    if job_designation:
        job_designation = job_designation.split(' at ')[0]  # Remove company information
        
        skills = profile.get('skills', [])  # Assuming skills is a list in the JSON structure
        
        job_designation_list.append(job_designation)
        skills_list.append(skills)

# Create a Pandas DataFrame
data = {'job_designation': job_designation_list, 'skills': skills_list}
df = pd.DataFrame(data)

# Display a sample of the DataFrame
df.sample(10)
```

```python
# Count the number of rows where 'skills' column contains null or empty lists
df['skills'].apply(lambda x: len(x) == 0).sum()
```
