```python
import pandas as pd
df = pd.read_excel('/kaggle/input/all-minilm-norm-dataset/all-miniLM-L12-v2_norm-dataset.xlsx')
```

```python
df.head()
```

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

# Encoding 'norm_profile' column
df['encoded_norm_profile'] = encoder.fit_transform(df['norm_profile'])

# Encoding 'norm_designation' column
df['encoded_norm_designation'] = encoder.fit_transform(df['norm_designation'])
```

```python
df.sample(10)
```

```python
# Reorder columns
columns_reordered = ['status', 'experience', 'present_ctc', 'expected_ctc', 'notice_period',
                     'candidate_profiles','norm_profile','encoded_norm_profile', 'job_categories', 
                     'designation', 'norm_designation','encoded_norm_designation','Relevant', 'exp_min', 'exp_max', 'ctc_fixed_min', 
                     'ctc_fixed_max']

# Reindex the DataFrame with the reordered columns
df = df.reindex(columns=columns_reordered)

df.head()  # Displaying the reordered DataFrame
```

```python
def fix_ctc_values(df1):
    def fix_ctc_lpa(x):
        if x > 1000:
            return x/100000
        else:
            return x

    df['expected_ctc'] = df['expected_ctc'].apply(fix_ctc_lpa)
    df['present_ctc'] = df['present_ctc'].apply(fix_ctc_lpa)
    return df1

fix_ctc_values(df)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Define the numeric features and scale them
numeric_features = ['experience', 'present_ctc', 'expected_ctc', 'notice_period',
                    'exp_min', 'exp_max', 'ctc_fixed_min', 'ctc_fixed_max']

scaler = StandardScaler()
scaled_numeric_data = scaler.fit_transform(df[numeric_features])
scaled_df = pd.DataFrame(scaled_numeric_data, columns=numeric_features)

# Select non-numeric features for concatenation
non_numeric_features = ['status', 'candidate_profiles', 'norm_profile', 
                        'encoded_norm_profile','job_categories',
                        'designation', 'norm_designation', 
                        'encoded_norm_designation','Relevant']

# Concatenate the scaled numeric data with the remaining columns
final_df = pd.concat([scaled_df, df[non_numeric_features]], axis=1)

# Display the updated DataFrame
final_df.head()
```

```python
# Reorder the columns as specified
desired_columns_order = ['status', 'experience', 'present_ctc', 'expected_ctc', 'notice_period',
                         'candidate_profiles', 'norm_profile', 'encoded_norm_profile',
                         'job_categories', 'designation', 'norm_designation', 'encoded_norm_designation',
                         'Relevant', 'exp_min', 'exp_max', 'ctc_fixed_min', 'ctc_fixed_max']

# Rearrange the columns of final_df
final_df = final_df[desired_columns_order]

# Display the DataFrame with reordered columns
final_df.head()
```

```python
# Encode the target variable into numeric format
le = LabelEncoder()
y = df['status']
y = le.fit_transform(y)

# Reverse the encoding
y = 1 - y

# Reshape y as a 1-dimensional array
y = y.ravel()
```

```python
from xgboost import XGBClassifier

# Initialize the XGBoost classifier
xgb = XGBClassifier(n_estimators=3,
                         max_depth=5,
                         learning_rate=0.125,
                         random_state=0)

# Define the input features and output label
input_features = ['experience', 'present_ctc', 'expected_ctc',
                  'notice_period', 'encoded_norm_profile', 
                  'encoded_norm_designation', 'exp_min', 'exp_max', 
                  'ctc_fixed_min', 'ctc_fixed_max']

X = df[input_features]
```

```python
from sklearn.model_selection import cross_val_score

def mean_accuracy(clf, X, y, cv=10):
    scores = cross_val_score(clf, X, y, cv=cv, scoring='accuracy')
    mean_acc = scores.mean()
    return mean_acc

mean_accuracy(xgb, X, y, cv=10)
```

```python
# Filter the DataFrame for rows where status is 'ACCEPTED'
accepted_profiles = df[df['status'] == 'ACCEPTED']

# Group by norm_designation and aggregate norm_profile values into a list
profiles_by_designation = accepted_profiles.groupby('norm_designation')['norm_profile'].agg(list).reset_index()

# Display the table
profiles_by_designation.head()
```

```python
# Assuming 'profiles_by_designation' contains the table with norm_profile against norm_designation

# Define the file path for the Excel file
file_path = 'profiles_by_designation.xlsx'

# Export the table to an Excel file
profiles_by_designation.to_excel(file_path, index=False)

print(f"Table exported to {file_path}")
```
